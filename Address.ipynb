```python
import re
import pandas as pd

#Reading the File
file = 'address.txt'
with open(file, encoding='utf8', errors='ignore') as f:
    l = f.readlines()
dd_data= [i.replace("\x00","").strip() for i in l]

#-----------------------------------------------------------
#Removing the Blank spaces from the Data
while("" in dd_data) : 
    dd_data.remove("") 
    
#Removing the Unwanted lines from the Data
for data in dd_data:
    if "Current Resident" in data:
        dd_data.remove(data)
    
names = []
address = []

#-----------------------------------------------------------
#Adding elements into the Names list
for i in range(0, len(dd_data), 3):
    names.append(dd_data[i])
    
#Joining all the List Elements    
result_string = '  '.join(names)
indi_names = result_string.split('  ')

#Printing  the individual Names
while("" in indi_names) : 
    indi_names.remove("") 

#-------------------------------------------------------------
#Adding elements into the Address list
for i in range(len(dd_data)):
    address.append(dd_data[i])
for i in range(0, len(dd_data), 3):
    address.remove(dd_data[i])
    

Address_line1 = []    
for i in range(0, len(address), 2):
    Address_line1.append(address[i])

Address_line2 = []
for i in range(1, len(address), 2):
    Address_line2.append(address[i])

result_string_line1 = '  '.join(Address_line1)
indi_line1 = result_string_line1.split('  ')

while("" in indi_line1) : 
    indi_line1.remove("") 
    
    
result_string_line2 = '  '.join(Address_line2)
indi_line2 = result_string_line2.split('  ')

while("" in indi_line2) : 
    indi_line2.remove("") 

#---------------------------------------------------------------
#Adding the Cities
Cities = []
for i in indi_line2:
    Cities.append(i.lstrip().split(",")[0])

#---------------------------------------------------------------
#Adding States and Zip
States_Zip = []
for i in indi_line2:
    States_Zip.append(i.lstrip().split(",")[1])

#Adding States
States = []

for i in States_Zip:
    States.append(i.split(" ")[1])
    
#----------------------------------------------------------------

#Adding ZipCodes
Zip = []
for i in States_Zip:
    Zip.append(i.split(" ")[2])

#------------------------------------------------------------------

#Adding the Names to the Column using the Data Frames
FinalData = pd.DataFrame({'Names': indi_names,
                        'Address': indi_line1,
                        'Cities': Cities,
                        'States': States,
                        'Zip':Zip
                        })

#------------------------------------------------------------------
#Generating an Excel file for the Final Data
FinalData.to_excel("Address.xlsx",sheet_name='Address')  
```
